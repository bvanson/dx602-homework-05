### Problem 10

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]

# of entries to extend Collatz sequence by
entries = 5

# get next entries
for i in range(entries):
    # get last entry in list
    last = q10[-1]
    
    #check whether even or odd 
    if (last % 2 == 0):
        q10.append(last/2)
    else:
        q10.append(3*last+1)

     
    
````

````python
# YOUR CHANGES HERE

plt.plot(q10,color = 'blue') # x values are implicit (0,1,2,....)
plt.xlabel('Index')
plt.ylabel('Collatz Value')
plt.title('Collatz Sequence (Starting from 15, 10 Entries)')
````

### Problem 18

Read the data from `f18.tsv` and parse it as a TSV file.
Divide the data by the "animal" column and in a subplot for each animal, plot the "age" and "weight" columns as the x and y axes respectively with a different subplot for each animal.
Each subplot should be a line chart, but add markers for each point as some animals only have one data point.
Make sure to mention the animal name in the chart titles, and label the axes.

````python
# YOUR CHANGES HERE
from matplotlib.pyplot import cm
import numpy as np
# This function checks if an animal is already in the dict
# If found, returns index of animal in dict, else None
def in_dict(dict,animal):
    i = 0
    for each_animal in dict:
        if (each_animal['animal'] == animal):
            return i
        else:
            i += 1
    return None

# Create a dictionary from f18.tsv
# Each entry in dict is of the form: animal, age list (x values), weight list (y values)

dict = []
with open('f18.tsv') as fp:
    header = fp.readline() # skip header line
  
    for line in fp:
        line = line.rstrip('\n').split('\t')

        # in_dict returns index of animal in dict if found, else None
        index = in_dict(dict,line[0])

        # if found in dict, append age and weight to lists, else create new dict entry
        if (index is not None):
            dict[index]['age'].append(float(line[1]))
            dict[index]['weight'].append(float(line[2]))
        else:
            dict.append({'animal': line[0], 'age':[ float(line[1]) ], 'weight':[float(line[2])]})


how_many_animals = len (dict)

# create a color list for plotting different animals in different colors
color = cm.rainbow(np.linspace(0, 1, how_many_animals))
color_list = [hue for hue in color]

# space out the subplots vertically
plt.figure(figsize=(4,14))

# create the subplots
for (i, each_animal) in enumerate(dict):
    plt.subplot(how_many_animals,1, i+1)
    plt.plot(each_animal['age'],each_animal['weight'], marker = '*', linestyle='dashed',color=color_list[i])

    # labels
    plt.title(each_animal['animal'].capitalize())
    for (x,y) in zip(each_animal['age'],each_animal['weight']):
        plt.annotate(f'{y}',xy = (x,y),ha = 'center') # annotate each point with its weight 
    plt.xlabel('Age')
    plt.ylabel('Weight')

# more spacing between subplots
plt.subplots_adjust(hspace=0.4, wspace=0.4)

````

Read the file `f20.tsv` and parse it as a TSV file.
Three of its columns are color, x, and y.
Plot these x, y values with the specified color using [`plt.scatter`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).


Hint: you may find it useful to separate the rows by color first, and use a separate call to `plt.scatter` for each color.

````python
# YOUR CHANGES HERE

# Function to check if color is already in dict
def in_dict(dict,color):
    for (i,each_row) in enumerate(dict):
        if (each_row['color'] == color):
            return i
    # Not in list
    return None

dict = [] # Start with empty dictionary
with open('f20.tsv') as fp:
    header = next(fp) # Skip header line
    
    for line in fp:
        line = line.rstrip('\n').split('\t')

        # in_dict returns index of color in dict if found, else None
        index = in_dict(dict,line[0])

        if (index is not None): # color fund in dict, add x to x values, y to y values
            dict[index]['x'].append(float(line[1]))
            dict[index]['y'].append(float(line[2]))
        else: # color not found in dict, add new entry
            dict.append({'color': line[0], 'x':[ float(line[1]) ], 'y':[ float(line[2] )]})
        
#loop through how many colors and plot eaach one 
how_many_colors = len (dict)

if (how_many_colors > 0): # there is data to plot
    plt.figure(figsize=(8,how_many_colors *2)) # space out the subplots vertically
    for each_color in dict:
        plt.scatter(each_color['x'],each_color['y'], color=each_color['color'], label=each_color['color'])
    plt.legend()
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title("Colored Scatter Plots Read from f20.tsv")
else:
    print("No data to plot")
````
